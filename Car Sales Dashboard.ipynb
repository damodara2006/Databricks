```
from pyspark.sql.functions import *
from pyspark.sql import *
```

```
df = spark.read.format('csv').option('header',True).option('inferSchema',True).load('/FileStore/tables/norway_new_car_sales_by_make.csv')
```

```
df = df.withColumnRenamed('Make',"Car_name")
```

```
df_Volkswagen = df.groupBy('Car_name','Year').agg(sum('Car_sold').alias('Total_sold')).orderBy(col('Total_sold').desc()).filter(col("Car_name")=='Volkswagen')
df_Volkswagen_pd = df_Volkswagen.toPandas()
df_Volkswagen_pd
```

```
import seaborn as sns
import matplotlib.pyplot as plt
```

```
plt.figure(figsize=(10, 8))
sns.barplot(x='Year', y='Total_sold', data=df_Volkswagen_pd)  # Increase width (default is 0.8)
plt.ylabel("Total Sold", fontsize=14)
plt.xlabel("Year", fontsize=14)
plt.title("Volkswagen Sales by Year in Norway", fontsize=16)
plt.xticks(rotation=45)
plt.show()

```

```
df_sold_car_2017 = df.filter(col('Year')== 2017 ).select('Car_name','Car_sold').dropDuplicates(subset=['Car_name']).orderBy(col('Car_sold').desc())
df_sold_car_2017 = df_sold_car_2017.toPandas()
```

```
plt.figure(figsize=(20, 20))
ax = sns.barplot(x='Car_sold', y='Car_name', data=df_sold_car_2017)

# Add values to the end of each bar
for p in ax.patches:
    pw = p.get_width() / 2
    ph = p.get_height()
    print(ph , p.get_y())
    y = p.get_y()  + ph /1.5
    ax.text(pw , y, int(pw), fontsize=18, color='black')

plt.xlabel("Cars Sold", fontsize=14)
plt.ylabel("Car Name", fontsize=14)
plt.title("Top Cars Sold in 2017", fontsize=18)
plt.show()
```
